```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Specify the file path
file_path = "/content/drive/MyDrive/Colab Notebooks/age_salary.csv"

# Read the CSV file into a DataFrame
age_salary_data = pd.read_csv(file_path)

# Extract features (X) and target variable (y)
X = age_salary_data[['Age', 'EstimatedSalary']]
y = age_salary_data['Purchased']


# Splitting the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)

# Scaling the features
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# Training a linear SVM model
svm_classifier = SVC(kernel='linear', random_state=0)
svm_classifier.fit(X_train, y_train)

# Making predictions on the test set
y_pred = svm_classifier.predict(X_test)


def plot_decision_boundary(X, y, classifier, title):
    h = .02  # Step size in the mesh

    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1

    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))

    Z = classifier.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.contourf(xx, yy, Z, cmap=plt.cm.Paired, alpha=0.8)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Paired)

    plt.title(title)
    plt.xlabel('Age')
    plt.ylabel('Estimated Salary')

    plt.show()

# Plotting decision boundary for training set
plot_decision_boundary(X_train, y_train, svm_classifier, 'SVM Decision Boundary (Training set)')

from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming y_test and y_pred are your actual and predicted labels
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")

```

```python
from sklearn.datasets import make_circles
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.metrics import classification_report
import matplotlib.pyplot as plt
import numpy as np

# Generating a synthetic dataset using make_circles
X_circles, y_circles = make_circles(n_samples=400, factor=0.5, noise=0.05, random_state=0)

# Splitting the dataset into training and testing sets
X_train_circles, X_test_circles, y_train_circles, y_test_circles = train_test_split(X_circles, y_circles, test_size=0.25, random_state=0)

# Scaling the features
scaler_circles = StandardScaler()
X_train_circles = scaler_circles.fit_transform(X_train_circles)
X_test_circles = scaler_circles.transform(X_test_circles)

# Training a linear SVM model on the new dataset
svm_classifier_circles = SVC(kernel='linear', random_state=0)
svm_classifier_circles.fit(X_train_circles, y_train_circles)

# Making predictions on the test set
y_pred_circles = svm_classifier_circles.predict(X_test_circles)


# Plotting decision boundary for the circles dataset
plot_decision_boundary(X_train_circles, y_train_circles, svm_classifier_circles, 'SVM Decision Boundary (Circles Training Set)')

# Returning the classification report for the circles dataset
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Assuming y_test and y_pred are your actual and predicted labels
accuracy = accuracy_score(y_test_circles, y_pred_circles)
precision = precision_score(y_test_circles, y_pred_circles)
recall = recall_score(y_test_circles, y_pred_circles)
f1 = f1_score(y_test_circles, y_pred_circles)

print(f"Accuracy: {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall: {recall:.4f}")
print(f"F1-score: {f1:.4f}")
```
